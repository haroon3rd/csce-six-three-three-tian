Using the gradient found above, write a short program to

```
# Implement your program here:

# Example program. You can modity/extend this.
#
# 1. define function 
# - this is an example (quadratic function) -- you should change it to the function given in problem 4)
# - also, you need to check for the termination condition, rather than using a fixed number of steps.
def E(w):
  return (w-2)*(w-4.5)*(w-5)*(w-5.5)
  np.power(w-2.5,2) + 1      # note: np.power() can take scalar or vector as argument.

# 2. define derivative function 
def dEdw(w):
  return 4*np.power(w,3)-51*np.power(w,2)+209.5*w-273.25

# 3. gradient descent 

w = 1     # initial value
eta = 0.02   # learning rate  ; try eta=0.8 and eta=0.001

n = 100

w_t = np.zeros(n)
Ew_t = np.zeros(n)

i=0
last_delta_w = 0
while i<n:
  w_t[i]=w
  Ew_t[i]=E(w)
  current_delta_w = dEdw(w)
  w = w - eta * current_delta_w     # gradient descent step
  
  # check breaking condition
  if np.abs(current_delta_w - last_delta_w) < 0.001:
    break

  last_delta_w = current_delta_w
  i +=1

#print(i)
#print(w_t[:i+1])

# 4. plot E(w) and the gradient descent steps 
x = np.linspace(1,7,50)
plt.plot(x,E(x),"-",w_t[:i+1],Ew_t[:i+1],"*-")
```

* two layers 5/10/30 hidden units code

```
# two layers 5/10/30 hidden units
# just need to change hidden_units = 5/10/30

import numpy as np
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Model

# config
eta  = 0.01          # learning rate 
num_epochs = 300
b_size = 4   # batch size
hidden_units = 30

# data
x_train = np.arange(-2*np.pi, 2*np.pi, 0.1)
y_train = np.sin(x_train)

x_test = x_train
y_test = y_train

x_extend = np.arange(-6*np.pi, 6*np.pi, 0.1)
y_extend = np.sin(x_extend)

# define model architecture
model = Sequential() # define feed-forward NN
model.add(Dense(hidden_units, input_dim=1, activation='tanh'))   # add hidden layer # Dense means fully connected, 2=# of neurons, input_dim=2 input units,
model.add(Dense(1, activation='tanh'))  # add output layer 

# optimizer
opt = keras.optimizers.Adam(learning_rate=eta)

# build model 
model.compile(loss='mean_squared_error',  # loss defines the error function E()
		optimizer=opt,
		metrics=['MeanAbsoluteError']) # metrics is the performance metric: may use 'accuracy' but the results will be against your intuition

# train & evaluate 
model.fit(x_train,y_train, epochs=num_epochs, batch_size=b_size)
score = model.evaluate(x_test, y_test, batch_size=b_size)
print('test loss (MSE), test MAE : ',score)
```

* Three layers: 5+3, 20+3 units cases code

```
# Three layers 5+3, 20+3 hidden units
# just need to change first_hidden_units = 5/20, second_hidden_units = 3

import numpy as np
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Model

# config
eta  = 0.01          # learning rate 
num_epochs = 300
b_size = 4   # batch size
first_hidden_units = 5
second_hidden_units = 3

# data
x_train = np.arange(-2*np.pi, 2*np.pi, 0.1)
y_train = np.sin(x_train)

x_test = x_train
y_test = y_train

x_extend = np.arange(-6*np.pi, 6*np.pi, 0.1)
y_extend = np.sin(x_extend)

# define model architecture
model = Sequential() # define feed-forward NN
model.add(Dense(first_hidden_units, input_dim=1, activation='tanh'))   # add hidden layer # Dense means fully connected, 2=# of neurons, input_dim=2 input units,
model.add(Dense(second_hidden_units, activation='tanh'))
model.add(Dense(1, activation='tanh'))  # add output layer 

# optimizer
opt = keras.optimizers.Adam(learning_rate=eta)

# build model 
model.compile(loss='mean_squared_error',  # loss defines the error function E()
		optimizer=opt,
		metrics=['MeanAbsoluteError']) # metrics is the performance metric: may use 'accuracy' but the results will be against your intuition

# train & evaluate 
model.fit(x_train,y_train, epochs=num_epochs, batch_size=b_size)
score = model.evaluate(x_test, y_test, batch_size=b_size)
print('test loss (MSE), test MAE : ',score)
```
